#### Analytic Hierarchy Process, AHP

```python
import numpy as np

RI = np.array([0, 1e-10, 0.52, 0.89, 1.12, 1.26, 1.36, 1.41, 1.46, 1.49, 1.52, 1.54, 1.56, 1.58, 1.59], dtype=float)
```

```python
A = np.array([[1, 2, 5], [1 / 2, 1, 2], [1 / 5, 1 / 2, 1]], dtype=float)
n = A.shape[0]
stand_A = A / np.sum(A, axis=0)  #归一化
```

##### 通过算术平均的方式计算权重

```python
w1 = np.sum(stand_A, axis=1) / n
print(f"期望权重 w = {w1}")
```

##### 通过特征值进行求解

```python
eig_value, eig_vectors = np.linalg.eig(A)
eig_value = np.real(eig_value)
index = np.argmax(eig_value)
w2 = eig_vectors[:, index] / np.sum(eig_vectors[:, index])
w2 = np.real(w2)
print(f"特征值权重 w = {w2}")
```

##### 计算一致性比例CR
$$
\begin{align}
C.I. &= \frac{\lambda_{max} - n}{n-1} \\
C.R. & = \frac{C.I.}{R.I.}
\end{align}
$$

```python
CI = (np.max(eig_value) - n) / (n - 1)
CR = CI / RI[n]
print(f"C.I. = {CI}, R.I. = {RI[n]} for n = {n}, C.R. = {CR}")
```

```python
if CR < 0.1:
    print("一致性检验通过")
else:
    print("判断矩阵无效")
```
